# Результаты работы алгоритма

Проиллюстрируем работу алгоритма:
1) intelligent_placer.processing_source_data() - Обработка исходных данных
- Исходное изображение
- Полученная маска
- Обрезка изображения по маске

2) intelligent_placer.processing_input_data() - Обработка входных данных
- Исходное изображение
- Полученная маска
- Результат размещения

3) intelligent_placer.check_image(path) - Алгоритм для проверки одной фотографии
- path - путь до изображения
- Исходное изображение
- Полученная маска
- Результат размещения


Установим необходимые модули:

```python
from intelligent_placer_lib import intelligent_placer
import matplotlib.pyplot as plt
```

Проведем анализ исходных предметов

```python
images = intelligent_placer.processing_source_data()
```

Выведем результат

```python
for image in images:
    
    fig, ax = plt.subplots(1, 3, figsize=(15, 6))
    ax[0].imshow(image[0], cmap='gray')
    ax[0].set_title('Input image')
    ax[1].imshow(image[1], cmap='gray')
    ax[1].set_title('Mask')
    ax[2].imshow(image[2], cmap='gray')
    ax[2].set_title('Image after mask')
```

Проведём анализ всех тестовых изображений

```python
images_input, placement_result = intelligent_placer.processing_input_data()
```

Выведем результат в формате [image name][case]-[result] и посмотрим на само размещение

```python
print(f"The result of a successful placement all unput data: {placement_result}")
for image in images_input:
    
    fig, ax = plt.subplots(1, 3, figsize=(15, 6))
    ax[0].imshow(image.origin, cmap='gray')
    ax[0].set_title('Input image')
    ax[1].imshow(image.mask, cmap='gray')
    ax[1].set_title('mask')
    ax[2].imshow(image.polygon, cmap='gray')
    ax[2].set_title('result')
    print(f"[{image.name}][{image.case.name}]-[{image.result}]")
```

Посмотрим на случаи неверной работы алгоритма, если таковые найдутся

```python
if placement_result == 1:
    print("All tests passed")
else:
    for image in images_input:
        if image.case.name != image.result:
                fig, ax = plt.subplots(1, 3, figsize=(15, 6))
                ax[0].imshow(image.origin, cmap='gray')
                ax[0].set_title('Input image')
                ax[1].imshow(image.mask, cmap='gray')
                ax[1].set_title('mask')
                ax[2].imshow(image.polygon, cmap='gray')
                ax[2].set_title('result')
                print(f"[{image.name}][{image.case.name}]-[{image.result}]")
```

Ниже представлена функция для проверки одной фотографии

```python
path = "test.jpg"
image = intelligent_placer.check_image(path)
```

Посмотрим на результат

```python
print("[image name][result]")
print(f"[{image.name}][{image.result}]")

fig, ax = plt.subplots(1, 3, figsize=(15, 6))
ax[0].imshow(image.origin, cmap='gray')
ax[0].set_title('Input image')
ax[1].imshow(image.mask, cmap='gray')
ax[1].set_title('mask')
ax[2].imshow(image.polygon, cmap='gray')
ax[2].set_title('result')
```
